```python
import pandas as pd
data = pd.read_csv("./btcusdt_data/btcusdt5m.csv")
data.head(20)
```

```python
predictors = ["EMA", "CMO","MINUSDM", "PLUSDM", "CLOSE", "CLOSEL1", "CLOSEL2", "CLOSEL3", "3O", "CMB"]

X = data[predictors]
y = data.UP_DOWN
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_test.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X_train[predictors[0:-2]]) 

X_train[predictors[0:-2]] = scaler.transform(X_train[predictors[0:-2]])

X_test[predictors[0:-2]] = scaler.transform(X_test[predictors[0:-2]])
X_test.head()
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=200)

# fit the model with data
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
T = cnf_matrix[0][0]+cnf_matrix[1][1]
F = cnf_matrix[0][1]+cnf_matrix[1][0]
print(cnf_matrix)
print(metrics.accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
